**Data Analytics and Convergence Trend**

```python
# Import Library
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib.patches import FancyBboxPatch
from matplotlib.ticker import ScalarFormatter
import matplotlib.scale as s

#Plot width / height
fig, ax = plt.subplots(figsize = (10,4.5), dpi=150)

#SUB PLOT  1
plt.subplot(1, 2, 1)

#plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.9, hspace=0.4)


# Define Data and labels
metrics = ['Accuracy','Precision','Loss']
spectrograms = [99.06, 100, 0.041]
raw_signals = [76.95, 80.29, 0.453]

x_axis = np.arange(len(metrics))


# Multi bar Chart

plt.bar(x_axis -0.2, spectrograms, width=0.4, label = 'Spectrograms',color='#FF7F0E')
plt.bar(x_axis +0.2, raw_signals, width=0.4, label = 'Raw Signals',color='#1F77B4')

# Xticks

plt.xticks(x_axis, metrics)


# Add labels

plt.xlabel("Performance Metrics", fontsize=14)
plt.ylabel("Statistics (Log Scale)", fontsize=14)
plt.title('Datasets Analytics')

# Add legend
plt.legend()


#Round the corners
new_patches = []
for patch in reversed(ax.patches): 
  bb = patch.get_bbox() 
  color=patch.get_facecolor() 
  p_bbox = FancyBboxPatch((bb.xmin, bb.ymin), abs(bb.width), abs(bb.height), boxstyle="round,pad=-0.0010,rounding_size=0.18", ec="none", fc=color, mutation_aspect=4 ) 
  patch.remove() 
  new_patches.append(p_bbox)
  

for patch in new_patches: ax.add_patch(patch)

plt.yscale('log')
plt.gca().yaxis.set_major_formatter(ScalarFormatter())

# Display

#plt.tight_layout(pad=3.0)
#plt.show()


#SUB PLOT 2
plt.subplot(1, 2, 2)

#Define data and labels
Value_sets = ['1','50','100','200','300','400','450','500']
spectrograms_Accuracies = [50.23,88.51,92.50,96.01,97.50,98.75,99.06,99.06]
rawSignals_Accuracies = [48.28,68.35,72.42,75.78,75.93,75.93,75.93,75.93]

#Plot the bars
plt.plot(Value_sets, rawSignals_Accuracies,label ='Raw Siganls',marker='o')
plt.plot(Value_sets, spectrograms_Accuracies,'-',label ='Spectrograms',marker='o')

#Plot labels
plt.title('Validation Accuracy')
plt.xlabel('No. of Epochs', fontsize=14)
plt.ylabel('Accuracy (%)', fontsize=14)

#define legend
plt.legend()

#labelling line data points
for i,j in zip(Value_sets,rawSignals_Accuracies):
    plt.annotate(str(j), (i, j),textcoords="offset points", # how to position the text
                 xytext=(0,-10), # distance from text to points (x,y)
                 ha='center',size=9.6)  #7.9

for i,j in zip(Value_sets,spectrograms_Accuracies):
    plt.annotate(str(j), (i, j),textcoords="offset points", # how to position the text
                 xytext=(0,3.0), # distance from text to points (x,y)
                 ha='center',size=10.1)  #8.3
    

# Display
plt.tight_layout(pad=2.5)
#plt.show()

#saving image
plt.savefig("/content/bar-line_datasetsAnalytics_Convergence.png",dpi=600, bbox_inches='tight',pad_inches = 0.12)
```
